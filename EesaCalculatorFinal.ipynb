```python
import tkinter as tk
from tkinter import *
from tkinter import messagebox



exp = ''
count = 0

def press(num): 
    global exp 
    exp = exp + str(num) 
    equat.set(exp) 
    
def clear():
    global exp
    exp = ''
    equat.set('')
    
def pressEqual():
    
    try:
        global exp
        result = str(eval(exp))
        equat.set(result)
        exp= ''
        
        
    except:
        equat.set('ERROR 404')
        exp = ''
def messages():
    MsgBox = tk.messagebox.showinfo('HELP','This is a Calculator of Various Purposes and Functions\nPress Bonus for Extra Features')

if __name__ =='__main__':
    root = tk.Tk()
    root.title('Eesa Calculator')
    root.configure(background = 'silver')
    root.geometry('350x250')
    equat = StringVar()
    exp_f = Entry(root,textvariable = equat)
    exp_f.grid(columnspan = 30, ipadx = 150)
    equat.set('Enter: ')
    
    one = Button(root,text= '1', fg = 'White', bg = 'blue', height = 2, 
                 width = 3, font = ('Arial', 12, 'bold'), command = lambda: press(1))
    one.grid(column = 0, row = 2)
    two = Button(root,text= '2', fg = 'White', bg = 'blue', height = 2, 
                 width = 3, font = ('Arial', 12, 'bold'), command = lambda: press(2))
    two.grid(column = 1, row = 2)
    three = Button(root,text= '3', fg = 'White', bg = 'blue', height = 2, 
                   width = 3, font = ('Arial', 12, 'bold'), command = lambda: press(3))
    three.grid(column = 2, row = 2)
    four = Button(root,text= '4', fg = 'White', bg = 'blue', height = 2, 
                  width = 3, font = ('Arial', 12, 'bold'), command = lambda: press(4))
    four.grid(column = 0, row = 3)
    five = Button(root,text= '5', fg = 'White', bg = 'blue', height = 2, 
                  width = 3, font = ('Arial', 12, 'bold'), command = lambda: press(5))
    five.grid(column = 1, row = 3)
    six = Button(root,text= '6', fg = 'White', bg = 'blue', height = 2, 
                 width = 3, font = ('Arial', 12, 'bold'), command = lambda: press(6))
    six.grid(column = 2, row = 3)
    seven = Button(root,text= '7', fg = 'White', bg = 'blue', height = 2, 
                   width = 3, font = ('Arial', 12, 'bold'), command = lambda: press(7))
    seven.grid(column = 0, row = 4)
    eight = Button(root,text= '8', fg = 'White', bg = 'blue', height = 2, 
                   width = 3, font = ('Arial', 12, 'bold'), command = lambda: press(8))
    eight.grid(column = 1, row = 4)
    nine = Button(root,text= '9', fg = 'White', bg = 'blue', height = 2, 
                  width = 3, font = ('Arial', 12, 'bold'), command = lambda: press(9))
    nine.grid(column = 2, row = 4)
    zero = Button(root,text= '0', fg = 'White', bg = 'blue', height = 2, 
                  width = 3, font = ('Arial', 12, 'bold'), command = lambda: press(0))
    zero.grid(column = 0, row = 5)
    equalBut = Button(root,text = '=', fg = 'blue', bg = 'silver', height = 2, 
                      width = 3, command = pressEqual,font = ('Arial', 12, 'bold'))
    equalBut.grid(column = 1, row = 5)
    plus = Button(root, text = '+', fg = 'white', bg = 'purple', height = 2,
                  width = 3, font = ('Arial', 12, 'bold'), command = lambda: press('+'))
    plus.grid(column = 3, row = 2)
    minus = Button(root, text = '-', fg = 'white', bg = 'purple', height = 2, 
                   width = 3, font = ('Arial', 12, 'bold'), command = lambda: press('-'))
    minus.grid(column = 3, row = 3)
    multiply = Button(root, text = 'x', fg = 'white', bg = 'purple', height = 2, 
                      width = 3, font = ('Arial', 12, 'bold'), command = lambda: press('*'))
    multiply.grid(column = 3, row = 4)
    divide = Button(root, text = '/', fg = 'white', bg = 'purple', height = 2, 
                    width = 3, font = ('Arial', 12, 'bold'), command = lambda: press('/'))
    decimalButton = Button(root, text = '.', fg = 'white', bg = 'purple', height = 2, 
                           width = 3, font = ('Arial', 12, 'bold'), command = lambda: press('.'))
    decimalButton.grid(column = 2, row = 5)
    clear = Button(root, text = 'C', fg = 'white', bg = 'purple', height = 2, 
                   width = 3, font = ('Arial', 12, 'bold'), command = clear)
    clear.grid(column = 3,row = 5)
    
    
    infoButton = Button(root, text = 'HELP', fg = 'white', bg = 'green', height = 2, 
                           width = 5, font = ('Arial', 12, 'bold'), command = messages)
    infoButton.grid(column = 15, row = 3) 
    exitButtonn = Button(root, text = 'EXIT', fg = 'white', bg = 'Red', height = 2, 
                           width = 5, font = ('Arial', 12, 'bold'), command = root.destroy)
    exitButtonn.grid(column = 15, row = 2)  

# Opening Up a New Canvas/Box
def openNew():
    if __name__ =='__main__':
    # Basic Information on New Canvas
        root = tk.Tk()
        root.title('Eesa Calculators')
        root.configure(background = 'silver')
        root.geometry('175x300')
        
    # Text Boxes being declared
        entry1 = tk.Entry (root) 
        entry2 = tk.Entry(root)
        taxEntry = tk.Entry(root)
        valueEntry = tk.Entry(root)
        percentageEntry = tk.Entry(root)
        
    # Placement of the objects
        entry1.grid(column = 0, row = 1)
        entry2.grid(column = 0, row = 2)
        taxEntry.grid(column = 0, row = 9)
        percentageEntry.grid(column = 0, row = 16)
        valueEntry.grid(column = 0, row = 17)
        
    # Gets the Sum of 2 numbers
        def getSum():
            x1 = entry1.get()
            x2 = entry2.get()

            label1 = tk.Label(root, text = float(x1)+float(x2))
            label1.grid(column = 1, row = 3)
        sumButton = Button(root, text = 'Sum', fg = 'black', bg = 'Pink', height = 2,
                        width = 7, font = ('Arial', 9, 'bold'), command  = getSum)
        sumButton.grid(column = 0, row = 3)   
        
    # Gets the product of 2 numbers
        def getProduct():
            x1 = entry1.get()
            x2 = entry2.get()

            label2 = tk.Label(root,text = float(x1) *float(x2))
            label2.grid(column = 1, row = 4)

        productButton = Button(root, text = 'Product', fg = 'black', bg = 'Pink', height = 2,
                            width = 7, font = ('Arial', 9, 'bold'), command  = getProduct)
        productButton.grid(column = 0, row = 4)

    # Button to get powers of two values
        def getPower():
            x1 = entry1.get()
            x2 = entry2.get()
            
            x = float(x1) ** float(x2)
            label3 = tk.Label(root,text = x)
            label3.grid(column = 1, row = 5)

        powerButton = Button(root, text = 'Power', fg = 'black', bg = 'Pink', height = 2,
                            width = 7, font = ('Arial', 9, 'bold'), command  = getPower)
        powerButton.grid(column = 0, row = 5)
    # Calculates 1/13 of given value
        def getTax():
            x1 = taxEntry.get()
            
            
            labelTax = tk.Label(root,text = float(x1)*1.13)
            labelTax.grid(column = 1, row = 10)
        taxButton = Button(root,text = 'Tax', fg = 'black', bg= 'Pink', height = 2,
                          width = 7, font = ('Arial',9, 'bold'), command = getTax)
        taxButton.grid(row = 15, column = 0)   
        
    # Calculates Percentage; First value is number while second is number   
        def getPercentage():
            perc = percentageEntry.get()
            val = valueEntry.get()
            
            percLabel = tk.Label(root, text = float(perc)*.01 * float(val))
            percLabel.grid(column = 1, row = 18)
            
        percButton = Button(root, height = 1, width = 9, command = getPercentage, text = 'Percentage',
                           fg = 'black', bg = 'pink', font = ('Arial', 9, 'bold'))
        percButton.grid(row = 18, column = 0)
    
bonusButton = Button(root, text = 'BONUS', fg = 'White', bg = 'Orange', height = 2,
                        width = 5, font = ('Arial', 12, 'bold'), command  = openNew)
bonusButton.grid(column = 15, row = 4)

#######################################################
######################################################
# TO DO LIST

def openList():
        # Enter An Empty Task List
        taskList = []
        # Counter
        count = 1
    
  # Function for Inserting a task
        def insertTask(): 
            global count 
            value = inpError() 
            if value == 0 : 
                return

            content = enterTaskField.get() + "\n"
            taskList.append(content) 
            textField.insert('end -1 chars', "[ " + str(count) + " ] " + content) 
            count += 1
            clearTasks() 
            
    # Function for Deleting a Task
        def delete() : 
            global count 
            if len(taskList) == 0 : 
                messagebox.showerror("No task") 
                return
            num = tasknumField.get(1.0, END) 
            if num == "\n" : 
                messagebox.showerror("input error") 
                return

            else : 
                task_no = int(num) 
            clearTaskNum() 
            taskList.pop(task_no - 1) 
            count -= 1
            textField.delete(1.0, END) 
            for i in range(len(taskList)) : 
                textField.insert('end -1 chars', "[ " + str(i + 1) + " ] " + taskList[i]) 
        
    # Function that checks potential errors
        def inpError() : 
            if enterTaskField.get() == "" : 
                messagebox.showerror("Error") 
                return 0

            return 1
    # Following Functions delete the tasks and num associated
        def clearTaskNum() : 
            tasknumField.delete(0.0, END) 
        def clearTasks() : 
            enterTaskField.delete(0, END) 
        
        if __name__ == "__main__" : 
        # Charactersitics of the Canvas -> Size, Name, Color ETC
            root = Tk()  
            root.configure(background = "Black") 
            root.title("To-Do List")  
            root.geometry("250x285") 
            
        # Enter Buttons, Labels and Boxes
            enterTask = Label(root, text = "Enter A Task", bg = "Purple", fg = 'White') 
            enterTaskField = Entry(root) 
            submitButton = Button(root, text = "Submit", fg = "White", bg = "Green", command = insertTask) 
            textField = Text(root, height = 5, width = 25, font = "Arial") 
            tasknum = Label(root, text = "Delete Task: ", bg = "Purple", fg = 'White') 
            tasknumField = Text(root, height = 1, width = 2, font = "Arial") 
            deleteButton = Button(root, text = "Delete", fg = "White", bg = "Green", command = delete) 
            exitButton = Button(root, text = "Exit", fg = "White", bg = "Red", command = root.destroy) 
            
        # Spacing of the objects
            enterTask.grid(row = 0, column = 2)                
            enterTaskField.grid(row = 1, column = 2, ipadx = 50) 
            submitButton.grid(row = 2, column = 2)  
            textField.grid(row = 3, column = 2, padx = 10, sticky = W) 
            tasknum.grid(row = 4, column = 2, pady = 5) 
            tasknumField.grid(row = 5, column = 2)                  
            deleteButton.grid(row = 6, column = 2, pady = 5) 
            exitButton.grid(row = 7, column = 2) 

 

         
listButton = Button(root,text = 'LIST', fg = 'white', bg = 'hot pink', height = 2, width = 5, font = ('Arial', 12, 'bold'),command = openList)
listButton.grid(column = 15, row = 5)

         
    
root.mainloop()
    
```
